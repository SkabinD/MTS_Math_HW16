# Подготовка

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
df = pd.read_csv("BankChurners.csv")
df.info()
df.describe()
```

# Основная часть

```python
plt.bar(df['Customer_Age'].value_counts().index, df['Customer_Age'].value_counts().values)
plt.show()
```

## Hist plot 

```python
sns.histplot(df, x="Customer_Age", binwidth=5, binrange=[25, 75])

plt.xticks(range(25, 76, 5))
plt.yticks(range(0, 2501, 100))

plt.xlim([25, 75])
plt.show()
```
